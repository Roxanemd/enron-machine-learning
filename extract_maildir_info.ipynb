```python
##########################"" Code for extracting info from mails

from nltk.stem.snowball import SnowballStemmer
import string
import re


def parseOutText(filename, colnames):
    """ given an opened email file f, parse out all text below the
        metadata block at the top
        (in Part 2, you will also add stemming capabilities)
        and return a string that contains all the words
        in the email (space-separated) 
        
        example use case:
        f = open("email_file_name.txt", "r")
        text = parseOutText(f)
        
        """

    f = open(filename, "r")
    f.seek(0)  ### go back to beginning of file (annoying)
    all_text = f.read()

    ### split off metadata
    content = all_text.split("X-FileName:")
    if len(content) > 0 : extractfromheader(filename,content[0], colnames)
    words = ""
    if len(content) > 1:
        ### remove punctuation
        text_string = content[1].translate(string.maketrans("", ""), string.punctuation)

        ### project part 2: comment out the line below
        words = text_string

        ### split the text string into individual words, stem each word,
        ### and append the stemmed word to words (make sure there's a single
        ### space between each stemmed word)
        
    return words

        
```

```python
import logging, sys
logging.basicConfig(stream=sys.stderr, level=logging.DEBUG)

# Fonction pour extraire les meta data du header
def extractfromheader(filename,header, colnames) :
    i = 0
    ###
    global result_dict
    header_lines = header.split('\n')
    # normal header length = 13 lines
    for line in xrange(0,len(header_lines)):
        if 'Message-ID:' in header_lines[line] : 
            result = re.search('<(.*)>', header_lines[0])
            v_msgid = result.group(1)
        elif 'Date:' in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[1])
            v_date = result.group(1).rstrip('\r\n')
        elif 'From:'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
        elif 'To:'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_to = result.group(1).rstrip('\r\n')
        elif 'Subject'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_subject = result.group(1).rstrip('\r\n')
        elif 'Mime-Version'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_mimvers = result.group(1).rstrip('\r\n')
        elif 'Content-Type'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_contenttype = result.group(1).rstrip('\r\n')
        elif 'Content-Transfer-Encoding'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_contenttypeencoding = result.group(1).rstrip('\r\n')
            #print "v_contenttypeeencoding : ",  v_contenttypeeencoding
        elif 'X-From'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_x_from = result.group(1).rstrip('\r\n')
        elif 'X-To'  in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_x_to = result.group(1).rstrip('\r\n')
        elif 'X-cc'   in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_x_cc = result.group(1).rstrip('\r\n')
        elif  'X-bcc'  in header_lines[line] :
            result = re.search('(.*)', header_lines[line])
            v_bcc = result.group(1).rstrip('\r\n')
        elif 'X-Folder'  in header_lines[line] :
            result = re.search('(.*)', header_lines[line])
            v_x_folder = result.group(1).rstrip('\r\n')
        elif 'X-Origin'   in header_lines[line] :
            result = re.search('\s(.*)\s', header_lines[line])
            v_x_origin = result.group(1).rstrip('\r\n')
        v_filename = filename 
    result_dict[v_msgid] = {}
    for col in colnames :
        if col != "v_msgid" :
            try :
                result_dict[v_msgid][col] = eval(col)
            except NameError:
                result_dict[v_msgid][col] = "null"
    #print result_dict[filename]
    
```
